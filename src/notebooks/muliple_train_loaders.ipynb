```python
from pytorch_lightning import LightningModule, Trainer
import torch
from torch.utils.data import DataLoader
from torch import nn
```

```python
class LitModel(LightningModule):
    def __init__(self):
        super().__init__()
        self.linear = nn.Linear(32, 32)
    def train_dataloader(self):

        loader_a = DataLoader(range(8), batch_size=4, shuffle=True)
        loader_b = DataLoader(range(16), batch_size=4, shuffle=True)
        loader_c = DataLoader(range(32), batch_size=4)
        loader_c = DataLoader(range(64), batch_size=4)

        # pass loaders as a nested dict. This will create batches like this:
#         loaders = {"loaders_a_b": [loader_a, loader_b], "loaders_c_d": {"c": loader_c, "d": loader_d}}
        loaders = [loader_a, loader_b]
        return loaders

    def training_step(self, batch, batch_idx):
        # access the data
        print(batch)
        
    def configure_optimizers(self):
        return torch.optim.Adam(self.parameters())
```

```python
model = LitModel()
```

```python
trainer = Trainer()
```
