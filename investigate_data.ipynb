```python
import numpy as np
```

```python
data = np.load('MLPopulation/data/Combined_GWTC_m1m2chieffz.npz')
```

```python
import seaborn as sns
import torch
```

```python
m1 = data['m1']
```

```python
m1 = torch.from_numpy(m1)
```

```python
def softplus_inv(y):
    return y + y.neg().expm1().neg().log()
```

```python
logm1 = softplus_inv(m1)
```

```python
logm1.mean()
```

```python
logm1.std()
```

```python
logm1_n = (logm1 - logm1.mean()) / logm1.std()
```

```python
sns.kdeplot(logm1_n[:, 0])
```

```python
sns.kdeplot(torch.nn.functional.softplus((logm1_n * logm1.std()) + logm1.mean())[:, 0])
```
